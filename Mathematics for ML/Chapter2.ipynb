# Manhattan Norm(1-norm) : $||x||_1 = \sum_{i=1}^{n} |x_i|$ 

The Manhattan Norm on $\mathbb{R}^2$ is the sum of the absolute values of the coordinates of the vector. It is also called the taxicab norm or the $L_1$ norm. 

```python
# For example:  
import numpy as np 

# Define a vector 

x = np.array([1, -2, 3, -4, 5]) 

# Calculate the L1 norm of the vector 
manhattan_norm = np.linalg.norm(x, ord=1) 

print("Vector: ", x)
print("L1 norm or Manhattan norm: ", manhattan_norm) 
```

# Applications of the Manhattan Norm

**1.Machine Learning and Data Science:**

* Feature Selection: The Manhattan norm is used in Lasso (Least Absolute Shrinkage and Selection Operator) regression, which is a type of linear regression that performs both variable selection and regularization to enhance the prediction accuracy and interpretability of the statistical model it produces.

* Sparse Representations: The (L_1) norm promotes sparsity in solutions, making it useful for problems where you want to enforce sparsity, such as in compressed sensing and sparse coding.

* Robustness to Outliers: The Manhattan norm is less sensitive to outliers compared to the Euclidean norm ((L_2) norm), making it useful in scenarios where robustness to outliers is important.

**2.Optimization:** 

* Linear Programming: The Manhattan norm is often used in linear programming problems where the objective is to minimize the sum of absolute differences.

* Regularization: The (L_1) norm is used as a regularization term in optimization problems to prevent overfitting by penalizing large coefficients.

**3.Computer Vision:**

* Image Processing: The Manhattan norm is used in image processing tasks, such as edge detection and image denoising, where the goal is to minimize the sum of absolute differences between pixel values.

**4.Robotics and Path Planning:**

* Pathfinding Algorithms:
The Manhattan norm is used in grid-based pathfinding algorithms, such as A* and Dijkstra's algorithm, to calculate the heuristic distance between points on a grid. This is particularly useful in environments where movement is restricted to horizontal and vertical directions, like in city street grids.

**5.Clustering:**

* K-Medians Clustering: The Manhattan norm is used in K-medians clustering, where the objective is to minimize the sum of absolute differences between data points and cluster centroids.

# Eucledian Norm(2-norm) : $||x||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}$ 

```python
# The Euclidean norm is the most commonly used norm. 
# It is also called the L2 norm. 
# It is defined as the square root of the sum of the squares of the elements of the vector. 
# It is given by the formula: 
# ||x||_2 = sqrt(x1^2 + x2^2 + ... + xn^2) 

import numpy as np 

# Define a vector 
x = np.array([1, -2, 3, -4, 5]) 

# Calculate the L2 norm of the vector 
euclidean_norm = np.linalg.norm(x, ord=2) 

print("Vector: ", x) 
print("L2 norm or Euclidean norm: ", euclidean_norm) 
```

# Applications of the Euclidean Norm

**1. Machine Learning and Data Science:**

* Distance Metrics: The Euclidean norm is commonly used to measure the distance between two points in Euclidean space. It is used in various machine learning algorithms, such as K-nearest neighbors (KNN) and clustering algorithms like K-means.

* Regularization: The Euclidean norm is used as a regularization term in ridge regression (L2 regularization) to prevent overfitting by penalizing large coefficients.

**2. Optimization:**

* Gradient Descent: The Euclidean norm is used to compute the magnitude of the gradient vector in optimization algorithms like gradient descent.

**3. Computer Vision:**

* Image Processing: The Euclidean norm is used to measure the difference between pixel values in image processing tasks, such as image denoising and edge detection.

**4. Robotics and Path Planning:**

* Pathfinding Algorithms: The Euclidean norm is used in pathfinding algorithms to calculate the straight-line distance between points, which is useful in environments where movement is not restricted to horizontal and vertical directions.

**5. Physics and Engineering:**

* Mechanics: The Euclidean norm is used to calculate the magnitude of vectors, such as force, velocity, and acceleration vectors in physics and engineering.



# Key Differences between L1 Norm and L2 Norm

**1. Definition:**

* **L1 Norm (Manhattan Norm):** The L1 norm is the sum of the absolute values of the vector elements.
    \[
    ||x||_1 = \sum_{i=1}^{n} |x_i|
    \]

* **L2 Norm (Euclidean Norm):** The L2 norm is the square root of the sum of the squares of the vector elements.
    \[
    ||x||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}
    \]

**2. Sensitivity to Outliers:**

* **L1 Norm:** The L1 norm is less sensitive to outliers compared to the L2 norm, making it more robust in scenarios where outliers are present.

* **L2 Norm:** The L2 norm is more sensitive to outliers because it squares the differences, giving more weight to larger differences.

**3. Sparsity:**

* **L1 Norm:** The L1 norm promotes sparsity in solutions, making it useful for feature selection and sparse representations.

* **L2 Norm:** The L2 norm does not promote sparsity and tends to produce solutions with non-zero coefficients.

**4. Use Cases:**

* **L1 Norm:** Used in Lasso regression, linear programming, and scenarios where robustness to outliers and sparsity are important.

* **L2 Norm:** Used in ridge regression, K-nearest neighbors, K-means clustering, and scenarios where the Euclidean distance is a natural measure of similarity.

**5. Example Values:**

* **L1 Norm of vector \( x \):** 15.0
* **L2 Norm of vector \( x \):** 7.416198487095663
