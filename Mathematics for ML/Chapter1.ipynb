When formalizing intuitive concepts a common approach is to construct a set of objects and a set of rules to manipulate these objects.

# Matrices

```python
# Matrices play a central role in linear algebra. They can be used to compactly represent systems of linear equations. 
# They can also be used to represent linear functions(linear mappings). 

# Definition: With m, n ∈ N, a matrix A with m rows and n columns is a rectangular array of numbers with m rows and n columns.
# Which is ordered according to a rectangular scheme consisting of m rows and n columns. 
# By convention (1,n) -matrices are called rows and (m,1)-matrices are called columns. These special matrices are also called row/column vectors 
# and are used to represent vectors in R^n.
```

# Linear mappings,
 
also known as linear transformations, are functions between two vector spaces that preserve the operations of vector addition and scalar multiplication. In other words, if ( T ) is a linear mapping from vector space ( V ) to vector space ( W ), then for any vectors ( \mathbf{u}, \mathbf{v} \in V ) and any scalar ( c ), the following properties hold:

Additivity: ( T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) )
Homogeneity: ( T(c \mathbf{u}) = c T(\mathbf{u}) )
These properties ensure that the structure of the vector space is preserved under the transformation.

```python
import numpy as np

# Define a linear transformation matrix
A = np.array([[2, 0], [0, 3]])

# Define a vector
v = np.array([1, 2])

# Apply the linear transformation
T_v = np.dot(A, v)

print("Original vector:", v)
print("Transformed vector:", T_v)
```

# Affine Mappings 

Affine mappings, or affine transformations, are functions between affine spaces that preserve points, straight lines, and planes. They are composed of a linear transformation followed by a translation. In other words, an affine mapping can be expressed as:

[ f(\mathbf{x}) = A\mathbf{x} + \mathbf{b} ]

where:

* ( A ) is a linear transformation matrix.

* ( \mathbf{x} ) is the input vector.

* ( \mathbf{b} ) is a translation vector.

```python
import numpy as np 
# Define linear transformation matrix 
A = np.array([[2, 0], [0, 3]]) 

# Define a translation vector 
b = np.array([1, 1]) 

# Define a vector 
v = np.array([1, 2]) 

# Apply the affine transformation
T_v = np.dot(A, v) + b 

print("Original vector:", v) 
print("Transformed vector:", T_v) 
```

# Key differences between linear and affine mappings: 


Translation Component:

Linear Mapping: 
* No translation component. The origin is always mapped to the origin.(Example: Scaling, rotation, reflection)

Affine Mapping:
* Includes a translation component. The origin can be mapped to any point.(Example: Translation, scaling with translation, rotation with translation)

Form:
Linear Mapping: 
* ( T(\mathbf{x}) = A\mathbf{x} )
Affine Mapping: 
* ( f(\mathbf{x}) = A\mathbf{x} + \mathbf{b} )

Preservation of Origin:

Linear Mapping: 
* Always maps the origin to the origin.

Affine Mapping: 
* Can map the origin to any point, depending on the translation vector ( \mathbf{b} ).

# ** Exercise Breakdown:**  
1. Show that ((\mathbb{R} \setminus {-1}, \star)) is an 
Abelian group:

Verify closure, associativity, identity element, inverse element, and commutativity.

2. Solve (3 \star x \star x = 15):

Simplify the given equation using the defined operation (\star).
Solve the resulting quadratic equation.

```python
import sympy as sp 

# Define the operations a ? b  := ab + a + b 
def star (a, b): 
    return a * b + a +b  

# Verify closure 
def verify_closure(a, b):  # Modified to accept two arguments
    result = star(a, b)
    return result != -1  

# Verify associativity 
def verify_associativity(a, b, c): 
    return star(star(a,b), c) == star(a, star(b,c))   

# Find identity element 
def  find_identity(): 
    e = sp.Symbol('e') 
    a = sp.Symbol('a') 
    eq = sp.Eq(star(a,e),a) 
    identity= sp.solve(eq, e) 
    return identity[0] 

# Find inverse element 
def find_inverse(a):
    b = sp.Symbol('b')
    e = find_identity() 
    eq = sp.Eq(star(a,b), e) 
    inverse = sp.solve(eq, b) 
    return inverse[0] 

# Verify commutativity 
def verify_commutativity(a, b):
    return star(a, b) == star(b, a) 

# Test values 
a, b, c = 2,3,4 

print("Closure:", verify_closure(a, b)) 
print("Associativity:", verify_associativity(a, b, c))
print("Identity element:", find_identity()) 
print("Inverse element:", find_inverse(a)) 
print("Commutativity:", verify_commutativity(a, b)) 
```

```python
# Define the symbols 
x = sp.Symbol('x') 

# Define the equation 3 ? x ? x = 15 
lhs = star(star(3, x), x) 
rhs = 15  
# Solve the equation 
solution = sp.solve(sp.Eq(lhs, rhs), x) 
print("Solution:", solution) 
```
