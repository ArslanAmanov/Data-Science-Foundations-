# Manhattan Norm(1-norm) : $||x||_1 = \sum_{i=1}^{n} |x_i|$ 

The Manhattan Norm on $\mathbb{R}^2$ is the sum of the absolute values of the coordinates of the vector. It is also called the taxicab norm or the $L_1$ norm. 

```python
# For example:  
import numpy as np 

# Define a vector 

x = np.array([1, -2, 3, -4, 5]) 

# Calculate the L1 norm of the vector 
manhattan_norm = np.linalg.norm(x, ord=1) 

print("Vector: ", x)
print("L1 norm or Manhattan norm: ", manhattan_norm) 
```

# Applications of the Manhattan Norm

**1.Machine Learning and Data Science:**

* Feature Selection: The Manhattan norm is used in Lasso (Least Absolute Shrinkage and Selection Operator) regression, which is a type of linear regression that performs both variable selection and regularization to enhance the prediction accuracy and interpretability of the statistical model it produces.

* Sparse Representations: The (L_1) norm promotes sparsity in solutions, making it useful for problems where you want to enforce sparsity, such as in compressed sensing and sparse coding.

* Robustness to Outliers: The Manhattan norm is less sensitive to outliers compared to the Euclidean norm ((L_2) norm), making it useful in scenarios where robustness to outliers is important.

**2.Optimization:** 

* Linear Programming: The Manhattan norm is often used in linear programming problems where the objective is to minimize the sum of absolute differences.

* Regularization: The (L_1) norm is used as a regularization term in optimization problems to prevent overfitting by penalizing large coefficients.

**3.Computer Vision:**

* Image Processing: The Manhattan norm is used in image processing tasks, such as edge detection and image denoising, where the goal is to minimize the sum of absolute differences between pixel values.

**4.Robotics and Path Planning:**

* Pathfinding Algorithms:
The Manhattan norm is used in grid-based pathfinding algorithms, such as A* and Dijkstra's algorithm, to calculate the heuristic distance between points on a grid. This is particularly useful in environments where movement is restricted to horizontal and vertical directions, like in city street grids.

**5.Clustering:**

* K-Medians Clustering: The Manhattan norm is used in K-medians clustering, where the objective is to minimize the sum of absolute differences between data points and cluster centroids.

# Eucledian Norm(2-norm) : $||x||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}$ 

```python
# The Euclidean norm is the most commonly used norm. 
# It is also called the L2 norm. 
# It is defined as the square root of the sum of the squares of the elements of the vector. 
# It is given by the formula: 
# ||x||_2 = sqrt(x1^2 + x2^2 + ... + xn^2) 

import numpy as np 

# Define a vector 
x = np.array([1, -2, 3, -4, 5]) 

# Calculate the L2 norm of the vector 
euclidean_norm = np.linalg.norm(x, ord=2) 

print("Vector: ", x) 
print("L2 norm or Euclidean norm: ", euclidean_norm) 
```

**4. Robotics and Path Planning:**

* Pathfinding Algorithms: The Euclidean norm is used in pathfinding algorithms to calculate the straight-line distance between points, which is useful in environments where movement is not restricted to horizontal and vertical directions.

**5. Physics and Engineering:**

* Mechanics: The Euclidean norm is used to calculate the magnitude of vectors, such as force, velocity, and acceleration vectors in physics and engineering.



# Key Differences between L1 Norm and L2 Norm

**1. Definition:**

* **L1 Norm (Manhattan Norm):** The L1 norm is the sum of the absolute values of the vector elements.
    \[
    ||x||_1 = \sum_{i=1}^{n} |x_i|
    \]

* **L2 Norm (Euclidean Norm):** The L2 norm is the square root of the sum of the squares of the vector elements.
    \[
    ||x||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}
    \]

**2. Sensitivity to Outliers:**

* **L1 Norm:** The L1 norm is less sensitive to outliers compared to the L2 norm, making it more robust in scenarios where outliers are present.

* **L2 Norm:** The L2 norm is more sensitive to outliers because it squares the differences, giving more weight to larger differences.

**3. Sparsity:**

* **L1 Norm:** The L1 norm promotes sparsity in solutions, making it useful for feature selection and sparse representations.

* **L2 Norm:** The L2 norm does not promote sparsity and tends to produce solutions with non-zero coefficients.

**4. Use Cases:**

* **L1 Norm:** Used in Lasso regression, linear programming, and scenarios where robustness to outliers and sparsity are important.

* **L2 Norm:** Used in ridge regression, K-nearest neighbors, K-means clustering, and scenarios where the Euclidean distance is a natural measure of similarity.

**5. Example Values:**

* **L1 Norm of vector \( x \):** 15.0
* **L2 Norm of vector \( x \):** 7.416198487095663

# Symmetric, Positive Definite Matrices

Symmetric and positive definite matrices are important concepts in linear algebra and with numerous applications in various fields, such as optimization, statistics, and machine learning and numerical analysis.

**1. Symmetric Matrices:** 
A matrix (A) is symmetric if it is equal to its transpose, i.e., A = A^T. Symmetric matrices have several important properties: 

**Properties of Symmetric Matrices:** 

 * Real Eigenvalues: Symmetric matrices have real eigenvalues, which makes them useful in various applications, such as principal component analysis (PCA) and spectral clustering. 

 * Orthogonal Eigenvectors: Eigenvectors corresponding to distinct eigenvalues of a symmetric matrix are orthogonal. This property simplifies the diagonalization of symmetric matrices and is used in the spectral theorem.

 * Orthogonal Diagonalization: Symmetric matrices can be orthogonally diagonalized, i.e., they can be decomposed into a diagonal matrix D and an orthogonal matrix P such that A = PDP^T. This property is used in the spectral theorem and the singular value decomposition (SVD).  
 
 * Positive Definiteness: Symmetric matrices can be classified based on their positive definiteness, which is a key property in optimization and numerical analysis. 
 A matrix (A) is positive definite if it is symmetric and all its eigenvalues are positive. Alternative, (A) is positive definite if for any non-zero vector(x): 
    \[
    x^TAx > 0
    \] 
    Positive definite matrices have several important properties: 

**Properties of Positive Definite Matrices:** 

* Convexity: Positive definite matrices are used in convex optimization problems, where the objective function is convex and has a unique minimum. 
    
* Quadratic Forms: Positive definite matrices are used to define quadratic forms, which are used in optimization, statistics, and machine learning to model relationships between variables. 
    
* Cholesky Decomposition: Positive definite matrices can be decomposed into the product of a lower triangular matrix and its transpose, known as the Cholesky decomposition. This decomposition is used in numerical methods to solve linear systems and compute matrix inverses. 
    
* Positive Semidefiniteness: A matrix (A) is positive semidefinite if it is symmetric and all its eigenvalues are non-negative. Positive semidefinite matrices have similar properties to positive definite matrices but allow for zero eigenvalues.

**Applications** 

* **1.Optimization:** Positive definite matrices are used in quadratic programming and other optimization problems to ensure convexity.


* **2.Machine Learning:** Covariance matrices in multivariate statistics and machine learning are often positive definite.


* **3.Numerical Analysis:** Positive definite matrices are used in numerical methods, such as the Conjugate Gradient method, for solving linear systems.

```python
import numpy as np 

# Create a symmetric matrix 
A = np.array([[2, -1, 0], [-1, 2, -1], [0, -1, 2]]) 

# Check if the matrix is symmetric 
is_symmetric = np.allclose(A, A.T) 
print("Matrix A is symmetric: ", is_symmetric) 


# Check if the matrix is positive definite 
# A matrix is positive definite if all its eigenvalues are positive. 
eigenvalues = np.linalg.eigvals(A) 
is_positive_definite = np.all(eigenvalues > 0) 
print("Matrix A is positive definite: ", is_positive_definite) 

# Perform the Cholesky decomposition 
L = np.linalg.cholesky(A) 
print("Cholesky decomposition of A: \n", L) 
print("Reconstructed matrix A from L: \n", np.dot(L, L.T))  
```

# Remark:(Cauchy-Schwarz inequality).
 For inner product vector space V, 
 the Cauchy-Schwarz inequality states that for all x, y in V, we have: 
 
 - |(x, y)| <= ||x|| * ||y||.

```python
# Sample 

import numpy as np 

# Define two vectors 
x = np.array([1, 2, 3]) 
y = np.array([4, 5, 6]) 

# Calculate the inner product 
inner_product = np.dot(x, y)  

# Calculate the Euclidean norm of x 
norm_x = np.linalg.norm(x) 
norm_y = np.linalg.norm(y) 

# Verify the Cauchy-Schwarz inequality 
cauchy_schwarz_inequality = inner_product <= norm_x * norm_y 

print("Vector x: ", x) 
print("Vector y: ", y) 
print("Inner product of x and y: ", inner_product) 
print("Euclidean norm of x: ", norm_x)
print("Euclidean norm of y: ", norm_y) 
print("Cauchy-Schwarz inequality holds: ", cauchy_schwarz_inequality)  
```

# Angles and Orthogonality

Orthogonality is a fundamental concept in linear algebra that plays a crucial role in various applications, such as signal processing, optimization, and machine learning. 

It refers to the relationship between two vectors when their inner product is zero. 
Orthogonal vectors are perpendicular to each other in geometric sense.

# Properties of Orthogonal Vectors
- **1.Perpendicularity:** In Euclidean space, orthogonal vectors are perpendicular to each other.
- **2.Zero Inner Product:** The inner product of orthogonal vectors is zero.
- **3.Pythagorean Theorem:** If ( \mathbf{u} ) and ( \mathbf{v} ) are orthogonal, then the Pythagorean theorem holds: [ |\mathbf{u} + \mathbf{v}|^2 = |\mathbf{u}|^2 + |\mathbf{v}|^2 ]
- **4.Orthogonal Sets:** A set of vectors is orthogonal if every pair of distinct vectors in the set is orthogonal.

```python
import numpy as np 

# Define two orthogonal vectors 
u = np.array([1, 0, 0]) 
v = np.array([0, 1, 0]) 

# Calculate the inner product 
inner_product = np.dot(u, v) 

# Check orthogonality
is_orthogonal = inner_product == 0 

print("Vector u: ", u) 
print("Vector v: ", v) 
print("Inner product of u and v: ", inner_product) 
print("Vectors u and v are orthogonal: ", is_orthogonal) 
```

# Orthonormal Basis

```python
# We characterized properties of basis vectors and found that in an n-dimensional vector space, 
# we need n basis vectors, i.e., n vectors that are linearly independent. 
# we used inner products to compute the length of vectors and angle between vectors. 
# In the following, we will discuss the special case where the basis vectors 
# are orthogonal to each other and where the length of each basis vector is 1. 
# Such a set of basis vectors is called an orthonormal basis. 
```

```python
import numpy as np 

# Define the standard basis vectors in R^3 
e1 = np.array([1, 0, 0]) 
e2 = np.array([0, 1, 0]) 
e3 = np.array([0, 0, 1]) 

# Function to check if vectors are orthonormal 
def check_orthonormal(vectors): 
    for i in range(len(vectors)):
        for j in range(len(vectors)): 
            inner_product = np.dot(vectors[i], vectors[j])
            if i == j:
                #check if the norm is 1 
                if not np.isclose(inner_product,1):
                    return False 
            else:
                #check if the inner product is 0 
                if not np.isclose(inner_product, 0):
                    return False 
    return True 

# List of basis vectors 
basis_vectors = [e1, e2, e3] 

# Check if the basis vectors are orthonormal 
is_orthonormal = check_orthonormal(basis_vectors) 

print("Basis vectors: ") 
print("e1: ", e1) 
print("e2: ", e2) 
print("e3: ", e3) 
print("The basis vectors are orthonormal: ", is_orthonormal) 
```
